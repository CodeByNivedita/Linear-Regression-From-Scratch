```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('price', axis=1).values  # all columns except target
y = df['price'].values.reshape(-1, 1)  # target

# Initialize scaler
scaler = StandardScaler()

# Fit on features and transform
X_scaled = scaler.fit_transform(X)

# Check
print("First 5 rows of scaled features:")
print(X_scaled[:5])
```

```python
import numpy as np

n_features = X_scaled.shape[1]
theta = np.zeros((n_features, 1))  # weights
bias = 0  # intercept
```

```python
learning_rate = 0.01
n_iterations = 1000
m = X_scaled.shape[0]  # number of samples
```

```python
for i in range(n_iterations):
    # Predictions
    y_pred = np.dot(X_scaled, theta) + bias
    
    # Compute error
    error = y_pred - y
    
    # Gradients
    dtheta = (1/m) * np.dot(X_scaled.T, error)
    dbias = (1/m) * np.sum(error)
    
    # Update weights
    theta -= learning_rate * dtheta
    bias -= learning_rate * dbias
    
    # Print loss every 100 iterations
    if i % 100 == 0:
        loss = (1/(2*m)) * np.sum(error**2)
        print(f"Iteration {i}, Loss: {loss:.2f}")
```

```python
# Final predictions
y_pred_final = np.dot(X_scaled, theta) + bias

# R^2 score
ss_total = np.sum((y - np.mean(y))**2)
ss_res = np.sum((y - y_pred_final)**2)
r2_score = 1 - (ss_res / ss_total)
print("R^2 Score:", r2_score)
```
